## Квартили распределения и график box-plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANGE = 30 + 1

sample = np.random.randint(10, 10 + RANGE, size=100)
sample_df = pd.DataFrame(sample)
print(sample)

plt.hist(sample, bins=RANGE)
plt.show()
```

### Квантили и квартили
Квантили распределения – значения признака, делящие распределение на некоторое число равных частей.  
Квартили – три точки (значения признака), которые делят упорядоченное множество данных на 4 равных части.

```python
print(np.quantile(sample, 0.25))
print(np.quantile(sample, 0.50))
print(np.quantile(sample, 0.75))
print(np.quantile(sample, 1))
```

```python
print(sample_df.quantile(q=0.25))
print(sample_df.quantile(q=0.50))
print(sample_df.quantile(q=0.75))
print(sample_df.quantile(q=1))
```

### Боксплот (график)  
Межквартильный размах (IQR) – разница между Q1 Q1 ﻿и Q3Q3﻿ (иногда используется в качестве меры центральной тенденции). Чем больше межквартильный размах – тем шире "ящик".  
Усы боксплота = 1.5∗IQR (полтора межквартильных расстояния).  
Значения, лежащие за усами обозначаются жирными точками.  

```python
import seaborn as sns

ax = sns.boxplot(y=sample)
ax = sns.swarmplot(y=sample, color="#00ffa6")
```

```python
sample_df.boxplot(patch_artist=True)
```
