## Нормальное распределение

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

SIZE = 1000
BINS = 100

normal_distribution = np.random.normal(loc=100, scale=50., size=SIZE)
plt.hist(normal_distribution, bins=BINS)
plt.show()
```

### Стандартизация (Z-преобразование)  
Это такое преобразование, которое позволяет любую шкалу перевести в стандартную Z-шкалу (Z-scores), где среднее значение будет равно нулю, а стандартное отклонение – равняться 1. Форма распределения при этом не изменится.

```python
import scipy.stats as stats

zscore = stats.zscore(normal_distribution)
plt.hist(zscore, bins=BINS)
plt.show()
```

### Задача
Считается, что значение IQ (уровень интеллекта) у людей имеет нормальное распределение со средним значением равным 100 и стандартным отклонением равным 15 (M = 100, sd = 15).  
Какой приблизительно процент людей обладает IQ > 125?

```python
mean = 100
sd = 15
IQ = 125

stats_norm = stats.norm(mean, sd)
plt.hist(stats_norm.rvs(size=SIZE), density=True, histtype='stepfilled', alpha=0.2)
plt.show()
```

```python
# cdf - Cumulative distribution function
print(f'На промежутке более 125 расположено: {(1 - stats_norm.cdf(IQ)) * 100:.2f}% значений')
```

Какой приблизительно процент людей обладает IQ  на промежутке от 70 до 112?

```python
print(f'На промежутке от 70 до 112 расположено: {(stats_norm.cdf(112) - stats_norm.cdf(70)):.2%} значений')

z1 = (70 - mean) / sd
z2 = (112 - mean) / sd
print(f'На промежутке от 70 до 112 расположено: {(stats.norm.cdf(z2) - stats.norm.cdf(z1)):.2%} значений')
```
