## Доверительный интервал (confidence interval)  
Это такой интервал в котором с определенной вероятностью (95%, 99%) находится среднее ГС.

```python
import numpy as np
import matplotlib.pyplot as plt

sample = np.random.normal(loc=100, scale=50., size=1000)
plt.hist(sample, bins=100)
plt.show()
```

```python
import scipy.stats as st

mean = np.mean(sample)
sd = np.std(sample, ddof=1)
n = len(sample)
se = sd / n ** 0.5

print((mean - st.norm.ppf(.975) * se, mean + st.norm.ppf(.975) * se)) # 0.975 = (1 + 0.95) / 2
```

```python
se = st.sem(sample)
print(st.t.interval(0.95, len(sample) - 1, loc=mean, scale=se))
```

```python
import statsmodels.stats.api as sms

print(sms.DescrStatsW(sample).tconfint_mean())
```
