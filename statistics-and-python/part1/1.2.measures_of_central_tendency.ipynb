## Меры центральной тенденции

### Гистограмма

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANGE = 20

sample = np.random.randint(RANGE, size=100)
sample_df = pd.DataFrame(sample)
print(sample)

plt.hist(sample, bins=RANGE)
plt.show()
```

```python
import seaborn as sns

sns.histplot(sample, bins=RANGE)
```

### Мода (mode)  
Зачение измеряемого признака, которое встречается максимально часто.

```python
print(max((sample.tolist().count(item), item) for item in set(sample.tolist())))
```

```python
import scipy.stats as stats

print(stats.mode(sample))
```

```python
print(sample_df.mode())
```

### Медиана (median)  
Значение признака, которое делит упорядоченное множество данных пополам.

```python
print(np.median(sample))
```

```python
print(sample_df.median())
```

### Среднее (mean, среднее арифметическое) 
Cумма всех значений измеренного признака, деленная на количество измеренных значений.

```python
print(np.mean(sample))
```

```python
print(sample_df.mean())
```
