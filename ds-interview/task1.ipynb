# Задача 1
Подкинули монету N раз. Кол-во случаев, когда выпал орёл, на 10% больше, чем кол-во случаев, когда выпала решка. При каком N мы можем сказать, что монета «нечестная» (орёл и решка выпадают с разной вероятностью)?

# Решение

Две гипотезы:  
H0: вероятность выпадения орла и решки одинакова;  
H1: вероятность не одинакова.  
Дано, что по результатам испытаний, количество случаев выпадения орла, на 10% больше.  

Пусть n - количество испытаний, x - количество случаев выпадения решки, тогда количество случаев выпадения орла y = 1.1x.  
n = х + 1.1*x;  
x = n/2.1;  
y = 1.1*n/2.1=11n/21  

Результаты подбрасывания монеты принимаем как биномиальное распределение.  
Воспользуемся биномиальным тестом библиотеки stats из scipy - binom_test, который позволит определить статистическую значимость отклонения от теоритически ожидаемого распределения (p = 0.5). Т.к. альтернативная гипотеза не двусторонняя, то установим параметр alternative='greater'.  

В качестве уровня значимости, при которо можно отвергнуть H0 гипотизу установим значение 0.05.

```python
from scipy import stats
n = 10 # начальное количество испытаний
p_val = 1.
while p_val >= 0.05:
    n += 1
    y = round(n*11/21)
    p_val = stats.binom_test(y, n, 0.5, alternative='greater')
print('Ответ, количество испытаний N = ' + str(n), \
      '(cлучаев выпадений орла = ' + str(x) + ', p-value = ' + str(p_val) + ')', sep='\n')
```
